```python
import pandas as pd 
import numpy as np 
data = ({
    "Student":["Alice","Bob","Charlie","David","Rohit","Agastya","Amar","Joy"],
    "Maths":[85,90,np.nan,92,88,95,80,91],
    "Science":[80,85,90,np.nan,88,92,np.nan,89],
    "English":[75,80,85,90,88,90,84,89],
    "Sanskrit":[70,75,np.nan,85,88,90,82,87],
    "Attendance":[90,85,95,80,88,np.nan,84,89]
})

df = pd.DataFrame(data)
df
```

#### isnull

```python
df.isnull().sum()
```

#### missing values without  mean

```python
Maths_avg = df["Maths"].sum() / df["Maths"].count()
df["Maths"] = df["Maths"].fillna(Maths_avg)
df
```

#### missing values verification

```python
df.isnull().sum()
```

#### remove rows more than 2 Nan value

```python
df = df[df.isnull().sum(axis=1) <= 2]
df
```
